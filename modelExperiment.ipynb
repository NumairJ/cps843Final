# Importing Lib

```python
from ultralytics import YOLO
```

## Build Models From YAML file, with predefined weights

```python
nano_model = YOLO("models/yolov5nu.pt")
#small_model = YOLO("yolov5s.yaml")
#medium_model = YOLO("yolov5m.yaml")
#large_model = YOLO("yolov5l.yaml")
#xlarge_model = YOLO("yolov5x.yaml")
```

```python
nano_model.info()
```

```python
results = nano_model.train(data="coco8.yaml", epochs=10, imgsz=640)
```

```python
results = nano_model.predict("datasets/coco8/images/val/000000000036.jpg")
# Print results
for r in results:
    print(r.boxes)  # bounding boxes
    print(r.boxes.xyxy)  # coordinates: x1, y1, x2, y2
    print(r.boxes.cls)   # class IDs
    print(r.boxes.conf)  # confidence scores

results[0].show()
```
